```python
# 1. Data Setup
# In this script, we will use simple OLS to

# 1.1 Importing the Python Modules we will use
# For this exercise, we will use the standard packages used when working on data in Python.  
import numpy as np
import pandas as pd

# This package allow us to get OLS tables like those in Excel, Stata and R. 
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col
```

```python
hiringData = pd.read_csv("Hyp_employees.csv")
hiringData.head()
```

```python
hiringData.shape
```

```python
hiringData.describe().T
```

```python
hiringData.corr()
```

# Example 1

```python
y = hiringData['mainperformancemetric']
```

```python
X = hiringData.drop(["mainperformancemetric", "new_id", "age"], axis=1)
```

```python
model1 = sm.OLS(y, X).fit()
```

```python
print(model1.summary())
```

```python
print("Table with Significance Stars")
print(summary_col(model1, stars=True))
```

```python
# 3.2 MBA/team quality exercise: didmba should have a positive effect on performance 
# if we don't include the team quality variable. But note that I cannot include
# outsideexperience either because otherwise MBA is not positive.

X_5 = hiringData.drop(["mainperformancemetric", "new_id", "tenure", "teamquality"],axis=1)
```

```python
model5 = sm.OLS(y, X_5).fit()
print(model5.summary())
```

```python
# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("Table with Significance Stars")
print(summary_col(model5, stars=True))
```

```python
# But the effect goes away when we include the team quality variable
x_6 = hiringData.drop({"mainPerformanceMetric","ID","outsideExperience"},axis=1)
model6 = sm.OLS(y,x_6).fit()
print(model6.summary())

# You can also produce a table with just the coefficients with significance 
# stars using the code below 
print("\n")
print("Table with Significance Stars")
print(summary_col(model6, stars=True))
```
